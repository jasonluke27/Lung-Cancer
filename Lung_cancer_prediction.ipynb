```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cancer patient data sets.csv")
df.head(10)
```

### Patient Id and index columns can be dropped

```python
df = df.drop(['index', 'Patient Id'], axis= 1)
df.head(10)
```

```python
values = df['Level'].value_counts()
print(values)
```

```python
vals = df['Level'].value_counts()
sns.barplot(x=vals.index, y=vals.values, hue=vals.index)
plt.show()
```

```python
def map_level_to_cancer(level):
    if level == 'High':
        return 1
    else:
        return 0

df['Cancer'] = df['Level'].apply(map_level_to_cancer)
```

```python
values = df['Cancer'].value_counts()
print(values)
```

### I have chosen to encode the 'Level' column into a binary classification column. This is because we will treat High as 1(Has Cancer) and Low as 0(Not Has Cancer). The Medium classification will go under 0(Not Has Cancer) because technically speaking the value is inconclusive enough to say the person has Lung Cancer. 

```python
df = df.drop('Level', axis=1)
```

## Feature Engineering

```python
df.columns
```

```python
feats = ['Age', 'Gender', 'Air Pollution', 'Alcohol use', 'Dust Allergy',
       'OccuPational Hazards', 'Genetic Risk', 'chronic Lung Disease',
       'Balanced Diet', 'Obesity', 'Smoking', 'Passive Smoker', 'Chest Pain',
       'Coughing of Blood', 'Fatigue', 'Weight Loss', 'Shortness of Breath',
       'Wheezing', 'Swallowing Difficulty', 'Clubbing of Finger Nails',
       'Frequent Cold', 'Dry Cough', 'Snoring']
targ = 'Cancer'

corr_matrix = df[feats + [targ]].corr()

# Extract correlation of features with the target variable
corr_target = corr_matrix[targ].drop(targ)

# Sort the correlations
corr_target_sorted = corr_target.sort_values(ascending=False)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_target_sorted.to_frame(), annot=True, cmap="Greens", cbar=False, fmt=".2f")
plt.title('Correlation of Features with Target (Cancer)')
plt.show()
```

#### From this heatmap we can drop coughing of blood, obesity, passive smoker and balanced diet as they have a very strong correlation with our Target variable.

```python
df = df.drop(['Coughing of Blood', 'Obesity', 'Passive Smoker', 'Balanced Diet'], axis=1)
```

```python
corr = df.corr()
plt.figure(figsize = (20,20))
sns.heatmap(corr, annot=True, cmap='Greens')
plt.show()
```

### When analysing all the features in our dataframe, we can see that Dust Allergy, Occupational Hazards and Genetic Risk have very high correlations with other features in the dataframe so it would be best to also drop these 3 features.

```python
df = df.drop(['Dust Allergy', 'OccuPational Hazards', 'Genetic Risk'], axis=1)
```

```python
corr = df.corr()
plt.figure(figsize = (15,10))
sns.heatmap(corr, annot=True, cmap='Greens')
plt.show()
```

### We will also drop the chronic lung disease, Alcohol use, Air Pollution and Smoking columns as they have strong correlations with other features in the data so it may overfit our model.

```python
df = df.drop(['chronic Lung Disease', 'Alcohol use', 'Air Pollution', 'Smoking'], axis=1)
```

```python
corr = df.corr()
plt.figure(figsize = (12,8))
sns.heatmap(corr, annot=True, cmap='Greens')
plt.show()
```

### Chest Pain and Fatigue also has a high correlation with our target variable so out of the 2 we can choose to drop Chest Pain as it has a higher correlation.

```python
df = df.drop(['Chest Pain'], axis=1)
```

# Model Training & Testing

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
```

### Feature Selection

```python
features = ['Age', 'Gender', 'Fatigue', 'Weight Loss', 'Shortness of Breath', 'Wheezing', 'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold',
'Dry Cough', 'Snoring']
target = 'Cancer'

X = df[features]
y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.5, random_state=42)
```

### We will need to oversample or undersample the data as we created a class count imbalance when encoding the cancer column. We will use undersampling to undersample class 1 to match class 0.

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(sampling_strategy=1)  # 1 means undersample to match the minority class

X_resampled, y_resampled = undersample.fit_resample(X_train, y_train)
```

#### When having only around 1000 samples to implement into our models, I have chosen to use a 50-50 split for the training and testing data.

## Logistic Regression

```python
lg = LogisticRegression(max_iter=500)
lg.fit(X_resampled, y_resampled)
y_pred = lg.predict(X_test)
classification = classification_report(y_test, y_pred)
print(classification)
```

## KNeighbors Classifier

```python
knn = KNeighborsClassifier()
knn.fit(X_resampled, y_resampled)
y_pred = knn.predict(X_test)
classification = classification_report(y_test, y_pred)
print(classification)
```

## XGBoost Classifier

```python
xgb = XGBClassifier()
xgb.fit(X_resampled, y_resampled)
y_pred = xgb.predict(X_test)
classification = classification_report(y_test, y_pred)
print(classification)
```
